```
df3 = pd.read_csv('covid+death.csv')
df3.head()
```

```
df3.tail()
```

```
df3.groupby(['Condition']).count()
```

Wordcloud where convert the column with diseases count into list using tolist() function

```
sentences = df3['Condition'].tolist()
```

converting the list into single string

```
sentences_as_string = ''.join(sentences)
```

convert the string into wordcloud

```
plt.figure(figsize=(14,14))
plt.imshow(WordCloud().generate(sentences_as_string))
```

#Converting condition group column into Wordcloud

convert column into tolist()

```
column2_tolist = df3['Condition Group'].tolist()
```

```
column_to_string = ''.join(column2_tolist)
```

```
plt.figure(figsize=(20,20))
plt.imshow(WordCloud().generate(column_to_string))
```
